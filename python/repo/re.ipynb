```python
import re
```

```python
'''
^ 匹配字符串的开头 | ^\d必须以数字开头 | $ 匹配字符串的末尾 | \d$必须以数字结尾
	^\d必须以数字开头, \d$必须以数字结尾
* 匹配0个或多个的字符 | + 匹配1个或多个的字符 | ? 匹配0个或1个字符 相当于{0,1} | . 匹配任意字符, 除了换行符\n, .. 匹配任意两个字符
	\(? 表示括号可有可无
	[) -]?  表示')'或者' '或者'-'或者什么都没有

\w 匹配一个字母或数字或下划线(\W非单词) | \s 匹配一个空格(也包括Tab等空白符)(\S匹配任意非空字符)
\D 非数字
\b 匹配一个单词边界(如, 'er\b'可以匹配"never"中的'er', 但不能匹配"verb"中的'er')(\B匹配非单词边界)
[\u4e00-\u9fa5]匹配中文 | [\u4e00-\u9fcc]匹配中文和中文符号 | [^abc] 匹配除了a, b, c以外的字符

{n} n个 | {n,} 如{1,}相当于x+
0\d{2,3} 0x或0xx   

正则匹配默认是贪婪匹配, 也就是匹配尽可能多的字符; 后加个?就可以采用非贪婪匹配
当?后缀于*,+,?,{n},{n,},{n,m}之后, 则代表非贪心匹配模式, 也就是说，尽可能少的匹配左边的字符或表达式

正则表达式有提取子串的功能, 用()表示的就是要提取的分组(Group)
m.group(0)是原始字符串, group(1)、group(2)...表示第1, 2, ...个子串
m.groups()返回所有分组(用括号括起, 默认是从右到左匹配, 若加了问号则从左到右)
'''
```

```python
# compile函数用于编译正则表达式, 生成一个正则表达式(Pattern)对象, 参数为模式字符串
pattern = re.compile(r'\d+') 
# re.compile(r'[^\u4e00-\u9fa5A-Za-z0-9]'), [\u4e00-\u9fa5]匹配中文, [^abc]匹配除了a, b, c以外的字符 
# re.compile(r'\w', re.UNICODE) # 处理任意字符
# r'[\w\u4e00-\u9fcc]+'

# match
# match()方法尝试从字符串的起始位置匹配一个模式, 如果匹配成功, 返回一个Match对象, 否则返回None
# re.match(r'^\d{3}\-\d{3,8}$', '010-12345')
m = pattern.match('one12twothree34four') # 查找头部, 没有匹配
print(m) # None
m = pattern.match('one12twothree34four', 2, 10) # 从'e'的位置开始匹配, 没有匹配
print(m) # None
m = pattern.match('one12twothree34four', 3, 10) # 从'1'的位置开始匹配, 正好匹配
print(m) # <_sre.SRE_Match object; span=(3, 5), match='12'>

print(re.match('www', 'www.runoob.com'))  # 在起始位置匹配, <_sre.SRE_Match object; span=(0, 3), match='www'>
print(re.match('com', 'www.runoob.com'))  # 不在起始位置匹配, None
# 正则表达式有提取子串的功能
print(re.match(r'^(\d{3})-(\d{3,8})$', '010-12345')) # <_sre.SRE_Match object; span=(0, 9), match='010-12345'>
print(re.match(r'^(\d{3})-(\d{3,8})$', '010-12345').groups()) # 返回元组, ('010', '12345')
print(re.match(r'^(\d{3})-(\d{3,8})$', '010-12345').group()) # 010-12345
print(re.match(r'^(\d{3})-(\d{3,8})$', '010-12345').group(0)) # 010-12345
print(re.match(r'^(\d{3})-(\d{3,8})$', '010-12345').group(1)) # 010
print(re.match(r'^(\d{3})-(\d{3,8})$', '010-12345').group(2)) # 12345

# findall
# re.findall(r'', text)返回一个列表
# match是匹配一次, findall是匹配所有
result1 = pattern.findall('runoob 123 google 456')
result2 = pattern.findall('run88oob123google456', 0, 10)
print(result1)
print(result2)
'''
['123', '456']
['88', '12']
'''

# search
# re.search(pat, s) 从字符串s的任意位置进行匹配
# 比如('123,'01234')就是匹配上了, 只要s存在符合pat的连续字符串就算匹配上了; 没有匹配上返回None, 匹配上返回matchobject
print(re.search('com', 'www.runoob.com')) # <_sre.SRE_Match object; span=(11, 14), match='com'>

# sub
# re.sub('', '', text) 将文本中的第一个参数替换成第二个参数, 与str.replace('', '')类似
# 如果第二个参数为str, 那么就是替换为第二个参数; 如果第二个参数是函数, 那么就按照函数返回值替换
# sub函数两个有默认值的参数分别是:
	# count: 表示最多只处理前几个匹配的字符串, 默认为0表示全部处理
	# flags: 默认为0
# 移除数字
# text = re.sub(r'\d+', '', text)
# 过滤评论中的表情
# pattern = re.compile(u'[\uD800-\uDBFF][\uDC00-\uDFFF]') # \u3000是空白字符
# print(pattern.sub('', text))
phone = "2004-959-559"
num = re.sub(r'\D', "", phone)
print("电话号码是 : ", num) # 电话号码是 :  2004959559
number = "123 4-567"
# \1...\9匹配第n个分组的内容; (?=..) 前向肯定界定符; (?!..) 前向否定界定符, 不包含..形式的串, 它是不占位置的, 起到一种指示作用
print(re.sub('(...?(?=..))', r'\1-', re.sub('\D', '', number)))
'''
123-45-67
'''
```
